# Feature scaling

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()

df.Amount=sc.fit_transform(pd.DataFrame(df.Amount))
df.head()
```

# Handling duplicates

```python
df=df.drop(['Time'],axis=1)
df.head()
```

```python
df.duplicated().any()
```

```python
df=df.drop_duplicates()
df.shape
```

# Seeing if imbalanced data or not

```python
df.Class.value_counts()
```

```python
import seaborn as sns
sns.countplot(x='Class', data=df)

```

# Splitting the data into training set and testing set

```python
x=df.drop('Class',axis=1)
y=df.Class
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.20, random_state=42, stratify=y)
```

# Undersampling

```python
from sklearn.utils import resample
new_df = pd.concat([x_train, y_train], axis=1)


normal = new_df[new_df['Class'] == 0]
fraud = new_df[new_df['Class'] == 1]
```

```python
normal_new = resample(normal,replace=False,n_samples=len(fraud),random_state=42)

undersampled_train = pd.concat([fraud, normal_new])
undersampled_train = undersampled_train.sample(frac=1, random_state=42)

x_train_under = undersampled_train.drop('Class', axis=1)
y_train_under = undersampled_train['Class']
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression(random_state=42)
log.fit(x_train_under,y_train_under)
```

```python
y_pred1=log.predict(x_test)
```

# Different evaluation metrics

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

# Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=42)
dt.fit(x_train_under,y_train_under)
```

```python
y_pred2=dt.predict(x_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
precision_score(y_test,y_pred2)
```

```python
f1_score(y_test,y_pred2)
```

# Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=42)
rf.fit(x_train_under,y_train_under)
```

```python
y_pred3=rf.predict(x_test)
```
